```python
from nltk.corpus import movie_reviews
import random

documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((list(movie_reviews.words(fileid)),category))

all_words = []
for x in movie_reviews.words():
    all_words.append(x)

all_words = nltk.FreqDist(all_words)


    
        
    
```

```python
word_features = list(all_words.keys())[:3000]
def find_features(document):
    features = {}
    wordset = set(document)
    for x in word_features:
        features[x] =(x in wordset)
    return features

print(find_features(documents[0][0]))


train_data = []

for doc in documents:
    train_data.append((find_features(doc[0]),doc[1]))


        
```

```python
print(len(train_data))
```

```python
random.shuffle(train_data)
train = train_data[:1900]
test = train_data[1900:]
classifier = nltk.NaiveBayesClassifier.train(train)
print("Accuracy :" ,(nltk.classify.accuracy(classifier , test)*100))
classifier.show_most_informative_features(20)

from nltk.classify.scikitlearn import SKlearnClassifier
```

```python
%%time
from nltk.corpus import state_union
text = state_union.raw("2005-GWBush.txt")
words = word_tokenize(text)
bigrams = ngrams(words,2)
trigrams = ngrams(words,3)
quadgrams = ngrams(words,4)
pentgrams = ngrams(words,5)

bi_dist = nltk.FreqDist(bigrams)
tri_dist = nltk.FreqDist(trigrams)
quad_dist = nltk.FreqDist(quadgrams)
pent_dist = nltk.FreqDist(pentgrams)

N_tri = sum(list(tri_dist.values()))
N_quad = sum(list(quad_dist.values()))
N_pent = sum(list(pent_dist.values()))


for x in tri_dist.keys():
    print("%s has probability : %.7f"%(x,tri_dist[x]/N_tri)) #Similarly for quad,pent
    
#lets find the probability of the trigram : Social Security Benefits
print("probability: %.7f"%(tri_dist[('Social','Security','benefits')] / bi_dist[('Social','Security')]   )   )
```

```python
%%time
from nltk.collocations import *
from nltk.metrics.association import QuadgramAssocMeasures
trigram_measures = nltk.collocations.TrigramAssocMeasures()
quadgram_measures = QuadgramAssocMeasures()

finder = QuadgramCollocationFinder.from_words(word_tokenize(text))

print("Quadgram PMI's:\n")
for i in finder.score_ngrams(quadgram_measures.pmi):
    print (i)
    
    
finder = TrigramCollocationFinder.from_words(word_tokenize(text))

print("Trigram PMI's:\n")
for i in finder.score_ngrams(trigram_measures.pmi):
    print (i)


```
